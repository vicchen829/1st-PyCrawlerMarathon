# ① 取出今天所有的發文

```python
url='https://www.ettoday.net/news/news-list.htm'
r=requests.get(url)
soup=BeautifulSoup(r.text,'html5lib')
```

```python
soup
```

```python
import datetime

today_yyyymmdd=datetime.datetime.today().strftime('%Y/%m/%d')

article_list=[]
```

```python
for d in soup.find(class_='part_list_2').find_all('h3'):
    
    date_time=d.find(class_='date').text
    date_time_yyyymmdd=datetime.datetime.strptime(date_time,'%Y/%m/%d %H:%M').strftime('%Y/%m/%d')
    
    category=d.find_all('em')[-1].text
    article=d.find_all('a')[-1].text
    
    if date_time_yyyymmdd==today_yyyymmdd:
        print(date_time,category,article)
        article_list.append([date_time,category,article])
```

# ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd
data=pd.DataFrame(article_list,columns=['日期','類別','標題'])
data
```

```python
data.groupby('類別').groups
```

# ③ 哪一個類別的文章最多

```python
df2=pd.DataFrame(data.groupby('類別').size(),columns=['文章數量'])
```

```python
df2
```

```python
df2.index
```

```python
df2['排名']=df2.rank(method='first',ascending=False)
```
