# ① 印出最新文章的「作者」「標題」「時間」

```python
import pandas as pd
url = 'https://www.ptt.cc/bbs/NBA/index.html'
r=requests.get(url)

soup=BeautifulSoup(r.text,'html5lib')
data=[]
```

```python
for d in soup.find_all(class_='title'):
    title=d.find_all('a')[-1].text
    print('標題',title)
    try:
        r2=BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text,'html5lib')
        author=r2.find_all(class_='article-meta-value')[0].text
        date_time=r2.find_all(class_='article-meta-value')[3].text
        print('作者：',author)
        print('時間：',date_time)
    except:
        continue
    
    data.append([title,author,date_time])
```

```python
df=pd.DataFrame(data,columns=['標題','作者','時間'])
```

```python
df
```

```python
def getdata(url):
    r=requests.get(url)

    soup=BeautifulSoup(r.text,'html5lib')
    data=[]
    
    for d in soup.find_all(class_='title'):
        if d.find_all('a')==[]: #本文已被刪除
            continue
        else:
            title=d.find_all('a')[-1].text
            try:
                r2=BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text,'html5lib')
                author=r2.find_all(class_='article-meta-value')[0].text
                date_time=r2.find_all(class_='article-meta-value')[3].text
            except:
                continue

            data.append([title,author,date_time])
    
    return pd.DataFrame(data,columns=['標題','作者','時間'])
```

# ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
my_df=getdata('https://www.ptt.cc/bbs/NBA/index.html')
```
