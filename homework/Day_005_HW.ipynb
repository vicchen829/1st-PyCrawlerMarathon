```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
p=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
p_json=json.loads(p.text)
```

```python
p_json
```

```python
print('1.ä¸€æ¬¡å›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿ Aï¼š',len(p_json))
print('2.æ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ')
for key in p_json[0].keys():
    print(key)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for p_jsons in p_json:
    print(p_jsons['title'])
    print(p_jsons['createdAt'])
    print(p_jsons['commentCount'])
    print(p_jsons['likeCount'])
    print('='*20)
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
hot_url='https://www.dcard.tw/_api/forums/pet/posts?popular=true'
nhot_url='https://www.dcard.tw/_api/forums/pet/posts?popular=false'

hot=json.loads(requests.get(hot_url).text)
nhot=json.loads(requests.get(nhot_url).text)
```

```python
print('ç†±é–€æ–‡ç« æ•¸',len(hot))
print('éç†±é–€æ–‡ç« æ•¸',len(hot))
```

```python
comment_num=0
like_num=0
for i in hot:
    comment_num+=i['commentCount']
    like_num+=i['likeCount']
print('ç†±é–€æ–‡ç« ï¼š\n')
print('ç¸½ç•™è¨€äººæ•¸=',comment_num,'æ–‡ç« æ•¸=',len(hot),'å¹³å‡ç•™è¨€äººæ•¸=',round(comment_num/len(hot),2))
print('ç¸½æŒ‰è®šäººæ•¸=',like_num,'æ–‡ç« æ•¸=',len(hot),'å¹³å‡æŒ‰è®šäººæ•¸=',round(like_num/len(hot),2))
```
