# 作業目標
請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：    

這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？    

取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」    
計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」    

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
p=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
p_json=json.loads(p.text)
```

```python
p_json
```

```python
print('1.一次回傳幾筆資料？ A：',len(p_json))
print('2.每一筆資料包含哪些欄位？')
for key in p_json[0].keys():
    print(key)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for p_jsons in p_json:
    print(p_jsons['title'])
    print(p_jsons['createdAt'])
    print(p_jsons['commentCount'])
    print(p_jsons['likeCount'])
    print('='*20)
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
hot_url='https://www.dcard.tw/_api/forums/pet/posts?popular=true'
nhot_url='https://www.dcard.tw/_api/forums/pet/posts?popular=false'

hot=json.loads(requests.get(hot_url).text)
nhot=json.loads(requests.get(nhot_url).text)
```

```python
print('熱門文章數',len(hot))
print('非熱門文章數',len(hot))
```

```python
comment_num=0
like_num=0
for i in hot:
    comment_num+=i['commentCount']
    like_num+=i['likeCount']
print('熱門文章：\n')
print('總留言人數=',comment_num,'文章數=',len(hot),'平均留言人數=',round(comment_num/len(hot),2))
print('總按讚人數=',like_num,'文章數=',len(hot),'平均按讚人數=',round(like_num/len(hot),2))
```

```python
comment_num=0
like_num=0
for i in nhot:
    comment_num+=i['commentCount']
    like_num+=i['likeCount']
print('熱門文章：\n')
print('總留言人數=',comment_num,'文章數=',len(nhot),'平均留言人數=',round(comment_num/len(nhot),2))
print('總按讚人數=',like_num,'文章數=',len(nhot),'平均按讚人數=',round(like_num/len(nhot),2))
```
